## Формат данных

Главная идея - соберем данные о переводах каждой из компаний в один вектор. 

Для каждой из компаний соберем три отдельные статистики - о переводах другим компаниям, о получении денег от других компаний и о платежах себе самой. 

Основная часть признаков, вычленяемых из статистик получения / отправки денег, практически идентична.

Для получения статистик используем _.groupby()_ отдельно по *hash_inn_kt*, *hash__inn__dt* и *hash__inn__kt* == *hash__inn__dt*

Hash каждой из компаний не используется

#### Функции для создания признаков

```python
hash_ = list(info['hash_inn'])
okved2 = list(info['okved2'])
region = list(info['region'])
hash_to_okved2 = dict(zip(hash_, okved2))
hash_to_region = dict(zip(hash_, region))

# в качестве признака сферы компании будем брать не само значение okved2, а его частоту в info.csv
freqs = list(info.groupby('okved2').count()['hash_inn'] / info.groupby('okved2').count()['hash_inn'].min())
l = [x for x in range(-1, 80, 1)]
okved_to_freq = dict(zip(l, freqs))
okved_to_freq[-1] = -1

# подсчитываем базовую статистику
def prim_features(type_, src_df):
    if type_ == 'sent' or type_ == 'both':
        index = 'hash_inn_kt'
    elif type_ == 'received':
        index = 'hash_inn_dt'
    
    src_df = src_df.sort_values(index)
    df = src_df.groupby(index).agg({'week': ['mean', 'median', 'std', 'min', 'max',],
                                    'count': ['mean', 'median', 'std', 'min', 'max'], # 'sum'
                                    'sum': ['mean', 'median', 'std', 'min'],}) # 'sum'
    return df


# Описание комплексных признаков (исп-ся в след. функции)
# mean_sent_transaction - среднее кол-во отправленных денег в неделю
# mean_receive_transaction - -//- полученных
# median_sent_transaction - медианное
# median_receive_transaction
# std_sent_transaction - ср.кв. отклонение
# std_receive_transaction
# num_sent - кол-во транзакций с отправлением
# num_received - -//- с получением
# num_sent_av_count - cр. кол-во транзакций с отправлением (по неделям)
# num_received_av_count - -//- с получением
# num_uniq_sent - кол-во уникальных компаний, которым переводились деньги
# num_uniq_received - кол-во уникальных компаний, от которых приходили деньги
# diff_week_sent - промежуток времени между первой и последней неделями покупки отправленных денег
# diff_week_received - промежуток времени между первой и последней неделями полученных денег
# diff_week_both
# diff_mean_sent_rec - разница отправленных и полученнных денег
# most_frequent_receiver_okved2 - самый частый получатель денег (по сфере деят.)
# least_frequent_receiver_okved2 - самый редкий
# most_frequent_sender_okved2
# least_frequent_sender_okved2 
# num_transactions_most_freq_sent - кол-во транзакций с самым частым контрагентом
# num_transactions_most_freq_rec
# -1..79_sent - сферы деят-ти компаний, которым отправлялись деньги
# -1..79_received - сферы деят-ти компаний, которые получали деньги
# mean_transaction_most_freq_sent - средняя сумма для самого частого контрагента
# mean_transaction_most_freq_rec
# max_transaction_most_freq_sent
# max_transaction_most_freq_rec
# min_transaction_most_freq_sent
# min_transaction_most_freq_rec


# type_: 
# sent - группируем pays.csv по компаниям которые отправляли деньги
# received - которые принимали
# both - которые отправляли себе же
def complex_features(x, type_):
    if type_ == 'sent' or type_ == 'both':
        index = 'hash_inn_dt'
        index_2 = 'receiver'
    elif type_ == 'received':
        index = 'hash_inn_kt'
        index_2 = 'sender'
    
    max_week = x['week'].max()
    min_week = x['week'].min()
  
    if type_ == 'sent' or type_ == 'received':
        # one-hot для индикации всех контрагентов
        l = [0] * 81
        ll = [str(x) + '_' + type_ for x in range(-1, 80, 1)]
        d = dict(zip(ll, l))
        # если сфера деят. компании -1 или она вообще неизвестна, добавляем 1 к флагу с индексом -1
        uniqs = list(x[index])
        for item in uniqs:
            try:
                d[str(hash_to_okved2[item]) + '_' + type_] = 1
            except KeyError:
                d[str(-1) + '_' + type_] = 1
        
        num_unique = x[index].nunique()
        most_freq = okved_to_freq[hash_to_okved2[x[index].value_counts().idxmax()]]
        least_freq = okved_to_freq[hash_to_okved2[x[index].value_counts().idxmin()]]
        
        temp_df_most_freq = x.loc[x[index] == most_freq]
        num_transactions_most_freq = temp_df_most_freq['count'].sum()
        mean_transaction_most_freq = temp_df_most_freq['sum'].mean()
        max_transaction_most_freq = temp_df_most_freq['sum'].max()
        min_transaction_most_freq = temp_df_most_freq['sum'].min()
        
        temp_df_least_freq = x.loc[x[index] == least_freq]
        num_transactions_least_freq = temp_df_least_freq['count'].sum()
        mean_transaction_least_freq = temp_df_least_freq['sum'].mean()
        max_transaction_least_freq = temp_df_least_freq['sum'].max()
        min_transaction_least_freq = temp_df_least_freq['sum'].min()
        
        feats = {'mean_' + type_ + '_transaction': (x['sum'] / x['count']).mean(),
                 'median_' + type_ + '_transaction': (x['sum'] / x['count']).median(),
                 'std_' + type_ + '_transaction': (x['sum'] / x['count']).std(),
                 'num_' + type_: x.shape[0],
                 'num_' + type_ + '_av_count': x['count'].mean(),
                 'num_unique_' + type_: num_unique,
                 'diff_week_' + type_: max_week - min_week,
                 'most_frequent_' + index_2 + '_okved2': most_freq,
                 'least_freq_' + index_2 + '_okved2': least_freq,
                 'num_transactions_most_freq_' + type_: num_transactions_most_freq,
                 'mean_transaction_most_freq_' + type_: mean_transaction_most_freq,
                 'max_transaction_most_freq_' + type_: max_transaction_most_freq,
                 'min_transaction_most_freq_' + type_: min_transaction_most_freq} 
        
        second = d
        return pd.Series({**feats, **second})
    
    elif type_ == 'both':
        feats = {'mean_' + type_ + '_transaction': (x['sum'] / x['count']).mean(),
                 'median_' + type_ + '_transaction': (x['sum'] / x['count']).median(),
                 'std_' + type_ + '_transaction': (x['sum'] / x['count']).std(),
                 'num_' + type_: x.shape[0],
                 'num_' + type_ + '_av_count': x['count'].mean(),
                 'diff_week_' + type_: max_week - min_week}

        return pd.Series(feats)


# признаки поверх получаемой таблицы
def final_features(src_df):
    src_df['diff_mean_sent_rec'] = src_df['mean_sent_transaction'] - src_df['mean_received_transaction']
    return src_df
```

Создаем признаки

```python
t0 = time()

# генерация признаков для тех, кто отправлял деньги
type_ = 'sent'
sent_df_prim = prim_features(type_, pays)
sent_df_compl = pays.groupby('hash_inn_kt').apply(complex_features, type_)
sent_df = pd.concat([sent_df_prim, sent_df_compl], axis=1)

# для тех тех, кто получал
type_ = 'received'
rec_df_prim = prim_features(type_, pays)
rec_df_compl = pays.groupby('hash_inn_dt').apply(complex_features, type_)
rec_df = pd.concat([rec_df_prim, rec_df_compl], axis=1)

# для тех, кто отправлял себе же
type_ = 'both'
temp_df = pays.loc[pays['hash_inn_kt'] == pays['hash_inn_dt']]
both_df_prim = prim_features(type_, temp_df)
both_df_compl = temp_df.groupby('hash_inn_kt').apply(complex_features, type_)
both_df = pd.concat([both_df_prim, both_df_compl], axis=1)

# соединяем все признаки
train_df = pd.concat([sent_df, rec_df, both_df], axis=1)

print('dataset creation time - {}'.format(time()-t0))
```

#### Удалим элементы, которых нет в info.csv

```python
train_df.index.name = 'hash_inn'
unique_hash = set(train_df.index)
info_set = set(info['hash_inn'])
to_del = list(unique_hash.difference(info_set))
num_deleted = 0
for item in to_del:
    train_df = train_df[train_df.index != item]
    num_deleted += 1
    
print('train_df.shape = {}, num deleted items = {}'.format(train_df.shape, num_deleted))
```

Заменим все NaN, появившиеся в результате слияния таблиц, на -1

```python
train_df = train_df.fillna(-1)
```

Добавим финальные признаки

```python
train_df = final_features(train_df)
train_df = train_df.fillna(-1)
```

Добавим таргет и регион

```python
hashes = list(train_df.index)
targets = [hash_to_okved2[x] for x in hashes]
regions = [hash_to_region[x] for x in hashes]
targets = pd.Series(targets, index=train_df.index)
regions = pd.Series(regions, index=train_df.index)
train_df['region'] = regions
train_df['target'] = targets
```

## Тренировка

```python
train_df = pd.read_csv('../tables/train_df.csv')
```

```python
path_data = '../tables/train_df_5.csv'
path_model = '../models/lgb_test.model'
```

```python
def run_lgb(train_X, train_y, test_X, test_y, val_X=None, val_y=None):
    params = {
        "objective" : "multiclass",
        "num_classes": 80,
        "num_threads": 4,
        "max_depth": 15, # 6
        "learning_rate" : 0.03,
#         "num_leaves" : 30,
        "early_stopping_round": 50,
        "metric" : "multi_logloss",
#         "min_child_samples" : 100,
        "verbosity" : -1,
        "bagging_seed" : 2018,
        "bagging_fraction" : 0.7,
        "feature_fraction" : 0.7,
        "num_iterations": 500
    }
    
    train = lgb.Dataset(train_X, label=train_y)
    test = lgb.Dataset(test_X, label=test_y)
    
    # можно использовать stratified k-folds, но не знаем, каким будет распределение в тестовом датасете
    
    verb_eval=5
    model = lgb.train(params, train, valid_sets=[test], # valid_sets=[train, val], 
                      valid_names = ['test'],
                      verbose_eval=verb_eval)
    return model


data = pd.read_csv(path_data)
data = data.sample(frac=1)
data.drop(['hash_inn'], 1, inplace=True)
data = data[data['target'] != -1]
train_cols = [x for x in list(data.columns) if x != 'target']

X = data[train_cols]
y = data['target']

p = 0.9
train_portion = int(X.shape[0] * p)
X_train, y_train = X.iloc[:train_portion], y.iloc[:train_portion]
X_test, y_test = X.iloc[train_portion:], y.iloc[train_portion:]

X_test.to_csv('X_test.csv')
y_test.to_csv('y_test.csv')

model = run_lgb(X_train, y_train, X_test, y_test)
model.save_model(path_model)
```
